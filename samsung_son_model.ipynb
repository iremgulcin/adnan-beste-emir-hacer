```python
import os

# KlasÃ¶rlerinizi kontrol edin
dataset_path = '/content/drive/My Drive/datason/'
print(os.listdir(dataset_path))
```

```python
from ultralytics import YOLO
```

```python
yaml_path = '/content/drive/My Drive/datason/dataa.yaml'
with open(yaml_path, 'r') as f:
    print(f.read())
```

```python
# YOLOv8'i temel alarak modeli baÅŸlat
model = YOLO('yolov8n.pt')  # 'n' versiyonu hÄ±zlÄ± ve hafif model iÃ§indir

# Modeli eÄŸit
model.train(
    data=yaml_path,        # YAML dosyasÄ±nÄ±n yolu
    epochs=50,             # EÄŸitim yapÄ±lacak epoch sayÄ±sÄ±
    imgsz=640,             # GÃ¶rÃ¼ntÃ¼ boyutu
    batch=16,              # Batch size
    name='fish_detection'  # Modelin adÄ±
)
```

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Modeli yÃ¼kle
model = YOLO('/content/runs/detect/fish_detection2/weights/best.pt')

# Test gÃ¶rÃ¼ntÃ¼sÃ¼nÃ¼ belirle
image_path = "/content/test.jpg"

# GÃ¶rÃ¼ntÃ¼yÃ¼ oku ve modelle tahmin yap
results = model(image_path)

# Tahmin edilen gÃ¶rÃ¼ntÃ¼yÃ¼ gÃ¶ster
res_plotted = results[0].plot()  # Ã‡izim yapÄ±lmÄ±ÅŸ sonucu al
plt.imshow(cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
